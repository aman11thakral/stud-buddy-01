```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, MultiLabelBinarizer, StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel("data40000.xlsx")

# Preprocess categorical and multi-label features
df['Skills'] = df['Skills'].apply(lambda x: x.split(', '))
df['Hobbies'] = df['Hobbies'].apply(lambda x: x.split(', '))

# One-hot encode categorical and binarize multi-label fields, excluding "Gender"
encoder = OneHotEncoder(sparse_output=False)
skills_binarizer = MultiLabelBinarizer()
hobbies_binarizer = MultiLabelBinarizer()

# Remove 'Gender' from the columns used for one-hot encoding
categorical_encoded = encoder.fit_transform(df[['Degree', 'Location']])  # Removed 'Gender'
skills_encoded = skills_binarizer.fit_transform(df['Skills'])
hobbies_encoded = hobbies_binarizer.fit_transform(df['Hobbies'])

# Normalize numerical features
scaler = StandardScaler()
age_scaled = scaler.fit_transform(df[['Age']])

# Combine all features, excluding 'Gender'
encoded_data = np.hstack([age_scaled, categorical_encoded, skills_encoded, hobbies_encoded])

# Reduce dimensionality with PCA
pca = PCA(n_components=20, random_state=42)  # Retain 20 components
pca_data = pca.fit_transform(encoded_data)

# Function to evaluate clustering and visualize results
def evaluate_clustering(data, labels, algo_name):
    silhouette = silhouette_score(data, labels)
    ch_score = calinski_harabasz_score(data, labels)
    db_score = davies_bouldin_score(data, labels)
    
    print(f"--- {algo_name} ---")
    print(f"Silhouette Score: {silhouette:.3f}")
    print(f"Calinski-Harabasz Score: {ch_score:.3f}")
    print(f"Davies-Bouldin Score: {db_score:.3f}\n")
    return silhouette

# Apply K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
df['Cluster_KMeans'] = kmeans.fit_predict(pca_data)
evaluate_clustering(pca_data, df['Cluster_KMeans'], "K-Means")

# Apply Gaussian Mixture Model
gmm = GaussianMixture(n_components=10, random_state=42)
df['Cluster_GMM'] = gmm.fit_predict(pca_data)
evaluate_clustering(pca_data, df['Cluster_GMM'], "GMM")

# Apply Spectral Clustering
spectral = SpectralClustering(n_clusters=10, affinity='nearest_neighbors', random_state=42)
df['Cluster_Spectral'] = spectral.fit_predict(pca_data)
evaluate_clustering(pca_data, df['Cluster_Spectral'], "Spectral Clustering")

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=2, min_samples=10)
df['Cluster_DBSCAN'] = dbscan.fit_predict(pca_data)
if len(np.unique(df['Cluster_DBSCAN'])) > 1:
    evaluate_clustering(pca_data, df['Cluster_DBSCAN'], "DBSCAN")
else:
    print("DBSCAN failed to form clusters with the current settings.\n")

# Visualize clusters with PCA
plt.figure(figsize=(12, 6))
for cluster_algo in ['Cluster_KMeans', 'Cluster_GMM', 'Cluster_Spectral']:
    plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df[cluster_algo], cmap='viridis', s=10, alpha=0.7, label=cluster_algo)
    plt.title(f"{cluster_algo} Visualization")
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.colorbar(label='Cluster')
    plt.show()

# Save the clustering results to Excel
output_path = "User_Profiles_Enhanced_Clustering_Excludi.xlsx"
df.to_excel(output_path, index=False)
print(f"Clustering results saved to {output_path}")
```

```python
import pandas as pd
import numpy as np

# Load the saved models
with open('kmeans_model.pkl', 'rb') as f:
    kmeans = pickle.load(f)
with open('pca_model.pkl', 'rb') as f:
    pca = pickle.load(f)
with open('encoder.pkl', 'rb') as f:
    encoder = pickle.load(f)
with open('skills_binarizer.pkl', 'rb') as f:
    skills_binarizer = pickle.load(f)
with open('hobbies_binarizer.pkl', 'rb') as f:
    hobbies_binarizer = pickle.load(f)
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

# Function to preprocess and predict
def predict_cluster(data):
    # Transform skills and hobbies
    data['Skills'] = data['Skills'].apply(lambda x: x.split(', '))
    data['Hobbies'] = data['Hobbies'].apply(lambda x: x.split(', '))
    
    # Encode categorical features
    categorical_encoded = encoder.transform(data[['Degree', 'Location']])
    skills_encoded = skills_binarizer.transform(data['Skills'])
    hobbies_encoded = hobbies_binarizer.transform(data['Hobbies'])
    
    # Normalize numerical features
    age_scaled = scaler.transform(data[['Age']])
    
    # Combine features
    encoded_data = np.hstack([age_scaled, categorical_encoded, skills_encoded, hobbies_encoded])
    
    # Dimensionality reduction
    pca_data = pca.transform(encoded_data)
    
    # Predict clusters using KMeans
    cluster = kmeans.predict(pca_data)
    return cluster
```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Parse the incoming JSON data
        user_data = request.json
        df = pd.DataFrame([user_data])
        
        # Predict cluster
        cluster = predict_cluster(df)
        return jsonify({'cluster': int(cluster[0])})
    
    except Exception as e:
        return jsonify({'error': str(e)}), 400

if __name__ == '__main__':
    app.run(debug=True)
```
